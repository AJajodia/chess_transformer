```python
# imports
import networkx as nx
import chess
```

```python
board = chess.Board(fen=None)
print(board)
print(list(chess.Board(fen=None).legal_moves))
```

```python


pieces = [chess.Piece.from_symbol(piece) for piece in ['R', 'K']]
G = nx.DiGraph()
nodes = [i for i in range(64)]
G.add_nodes_from(nodes)

for piece in pieces:
    for square in nodes:
        temp_board = chess.Board(fen=None)
        temp_board.set_piece_at(square, piece)
        moves = list(temp_board.legal_moves)
        for move in moves:
            s = chess.Move.uci(move)
            if chess.Piece.symbol(piece) == 'K':
                color = 'r'
            else:
                color = 'b'
            G.add_edge(square, chess.parse_square(s[-2:]), color = color, directed=True)

            
        
        
```

```python
colors = nx.get_edge_attributes(G,'color').values()
nx.draw(G, edge_color=colors, with_labels=True)
```
